```python
# Create Retail Sales Dataset

import pandas as pd

data = {
    'Customer_ID':[101,102,103,104,105,106,107,108,109,110],
    'Age':[22,35,45,29,41,38,26,50,33,27],
    'Gender':['Male','Female','Male','Female','Male','Female','Male','Female','Male','Female'],
    'Product':['Laptop','Mobile','Tablet','Laptop','Mobile','Tablet','Laptop','Mobile','Tablet','Laptop'],
    'Quantity':[1,2,1,3,2,1,2,1,3,2],
    'Price':[50000,20000,15000,50000,20000,15000,50000,20000,15000,50000]
}

df = pd.DataFrame(data)

print("Dataset Created Successfully")
df.head()
```

```python
# Dataset Information

print("Shape:", df.shape)
print("\nColumns:")
print(df.columns)

print("\nInfo:")
df.info()
```

```python
# Statistical Summary

print("Statistical Summary:")
df.describe()
```

```python
# Missing Values

print("Missing Values:")
print(df.isnull().sum())
```

```python
# Sales Distribution

import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df['Price'], bins=5)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# Product-wise Sales

df.groupby('Product')['Price'].sum().plot(kind='bar')

plt.title("Product Wise Sales")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Gender Distribution

df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')

plt.title("Gender Distribution")
plt.ylabel("")
plt.show()
```
